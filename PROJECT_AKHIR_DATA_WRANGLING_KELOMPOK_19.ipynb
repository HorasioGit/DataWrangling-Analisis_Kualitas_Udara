**TEKNIK PENGAMBILAN DAN INTEGRITAS**

```python
import pandas as pd
import numpy as np

file_UCI = 'PRSA_Data_Guanyuan_20130301-20170228.csv'
file_kaggle = 'air_pollution_china.csv'

df_prsa = pd.read_csv(file_UCI)
df_prsa
```

```python
df_kaggle = pd.read_csv(file_kaggle)
df_kaggle
```

```python
df_prsa_filtered = df_prsa[
    (df_prsa['year'] >= 2015) & (df_prsa['year'] <= 2017)
].copy()

df_prsa_filtered.head()
```

```python
df_kaggle_filtered = df_kaggle[
    (df_kaggle['Year'] >= 2015) & (df_kaggle['Year'] <= 2017)
].copy()

df_kaggle_filtered.head()
```

```python
agg_prsa = {
        'PM2.5': 'mean', 'PM10': 'mean', 'SO2': 'mean', 'NO2': 'mean',
        'CO': 'mean', 'O3': 'mean', 'TEMP': 'mean', 'PRES': 'mean',
        'DEWP': 'mean', 'WSPM': 'mean',
        'RAIN': 'sum',
        'wd': lambda x: x.mode()[0] if not x.mode().empty else np.nan
    }

df_prsa_monthly = df_prsa_filtered.groupby(
        ['year', 'month', 'station']
    ).agg(agg_prsa).reset_index()

df_prsa_monthly.rename(columns={'station': 'location'}, inplace=True)

df_prsa_monthly.head()
```

```python
rename_map_china = {
        'Year': 'year', 'Month': 'month', 'City': 'location',
        'PM2.5 (µg/m³)': 'PM2.5', 'PM10 (µg/m³)': 'PM10',
        'SO2 (µg/m³)': 'SO2', 'NO2 (µg/m³)': 'NO2', 'CO (mg/m³)': 'CO',
        'O3 (µg/m³)': 'O3', 'Temperature (°C)': 'TEMP',
        'Pressure (hPa)': 'PRES', 'Precipitation (mm)': 'RAIN',
        'Wind Speed (m/s)': 'WSPM', 'Wind Direction (°)': 'wd_deg',
        'Humidity (%)': 'Humidity', 'Visibility (km)': 'Visibility',
        'Day of Week': 'Day_of_Week', 'Weather Condition': 'Weather_Condition',
        'Station ID': 'Station_ID'

    }
df_kaggle_std = df_kaggle_filtered.rename(columns=rename_map_china)

cols_to_agg_kaggle = [
        'PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES',
        'WSPM', 'wd_deg', 'Humidity', 'Visibility', 'AQI', 'Latitude', 'Longitude'
    ]

agg_kaggle = {col: 'mean' for col in cols_to_agg_kaggle}
agg_kaggle['RAIN'] = 'sum'

df_kaggle_monthly = df_kaggle_std.groupby(
        ['year', 'month', 'location']
    ).agg(agg_kaggle).reset_index()

df_kaggle_monthly.head()
```

```python
output_filename = 'data_gabungan_bulanan_2015-2017.csv'

try:
  combined_monthly_df = pd.concat(
        [df_prsa_monthly, df_kaggle_monthly],
        ignore_index=True,
        sort=False
    )

  combined_monthly_df.to_csv(output_filename, index=False, float_format='%.2f')
  print(combined_monthly_df.head())

except FileNotFoundError as e:
    print(f"ERROR: File tidak ditemukan: {e.filename}")
    print("Pastikan nama file sudah benar dan ada di folder yang sama.")
except Exception as e:
    print(f"Terjadi error saat menjalankan skrip: {e}")
```

**Data Cleanning**

```python
import pandas as pd
```

```python
df = pd.read_csv("data_gabungan_bulanan_2015-2017.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
(df.isnull().mean() * 100)
```

```python
missing_percent = df.isnull().mean() * 100

df = df.loc[:, missing_percent < 50]

print("Kolom tersisa setelah menghapus yang missing > 50%:")
print(df.columns)
```

```python
# Isi numerik dengan median
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df[col].fillna(df[col].median(), inplace=True)

# Isi kategorikal dengan modus
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Cek apakah masih ada missing
df.isnull().sum()
```

```python
import numpy as np

# Gabungin waktu
df["date"] = pd.to_datetime(df["year"].astype(str) + "-" + df["month"].astype(str) + "-01")

df = df.drop_duplicates()

for col in df.columns:
    if col != "date":
        df[col] = pd.to_numeric(df[col], errors='ignore')

# Isi nilai kosong di kolom numerik dengan median
num_cols = df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Isi nilai kosong di kolom kategorikal dengan modus
obj_cols = df.select_dtypes(include=['object']).columns
for col in obj_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("Jumlah missing value setelah final cleaning:")
print(df.isnull().sum()[df.isnull().sum() > 0])

# Ubah kolom date jadi string
df["date"] = df["date"].dt.strftime("%Y-%m-%d")

df.to_csv("clean_data.csv", index=False)
df
```

**EDA (Exploratory Data Analysis)**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 4)

# Load file hasil cleaning
df = pd.read_csv("clean_data.csv")

# Pastikan kolom waktu benar
if 'date' in df.columns:
    df["date"] = pd.to_datetime(df["date"])

# Cek 5 data awal
df.head()
```

```python
print("Jumlah baris:", df.shape[0])
print("Jumlah kolom:", df.shape[1])
print("\nTipe data tiap kolom:")
print(df.dtypes)

print("\nStatistik deskriptif:")
display(df.describe())
```

```python
pollutants = ["PM2.5", "PM10", "SO2", "NO2", "CO", "O3"]

for col in pollutants:
    plt.figure()
    sns.histplot(df[col], kde=True, bins=25, color="steelblue")
    plt.title(f"Distribusi {col}")
    plt.xlabel(f"{col} (µg/m³)")
    plt.ylabel("Frekuensi")
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="location", y="PM2.5", palette="coolwarm")
plt.title("Perbandingan Tingkat PM2.5 antar Lokasi (2015-2017)")
plt.xticks(rotation=45)
plt.show()
```

```python
yearly_pm25 = df.groupby('year')['PM2.5'].mean().reset_index()
yearly_pm25["year"] = yearly_pm25["year"].astype(str)

plt.figure(figsize=(7,4))
sns.lineplot(data=yearly_pm25, x="year", y="PM2.5", marker="o", color="green")
plt.title("Tren Rata-rata PM2.5 per Tahun (2015–2017)")
plt.ylabel("Rata-rata PM2.5 (µg/m³)")
plt.xlabel("Tahun")
plt.grid(True)
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
corr = df[numeric_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap Korelasi Antar Variabel Numerik")
plt.show()
```

```python
df["TEMP_CAT"] = pd.cut(df["TEMP"], bins=3, labels=["Dingin", "Sedang", "Panas"])

plt.figure(figsize=(7,4))
sns.boxplot(data=df, x="TEMP_CAT", y="PM2.5", palette="coolwarm")
plt.title("Pengaruh Kategori Suhu terhadap PM2.5")
plt.xlabel("Kategori Suhu")
plt.ylabel("PM2.5")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(data=df.groupby("month")["PM2.5"].mean(), marker="o", color="orange")
plt.title("Rata-rata PM2.5 per Bulan (Musiman)")
plt.ylabel("PM2.5 (µg/m³)")
plt.xlabel("Bulan")
plt.show()
```
